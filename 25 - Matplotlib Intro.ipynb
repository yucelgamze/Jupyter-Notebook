```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [20,25,30,35,40,45,50,55,60]
```

```python
weight_list = [50,52,54,56,57,61,62,55,51]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy, weight_list_numpy, "Blue")
plt.title("Age Weight Graph")
plt.xlabel("Age")
plt.ylabel("Weight")
```

```python
my_numpy1 = np.linspace(0,15,20)
```

```python
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3
```

```python
my_numpy2
```

```python
plt.plot(my_numpy1,my_numpy2, "ro-")
```

```python
plt.subplot(1,2,1) # 1 row, 2 column, 1. graph
plt.plot(my_numpy1,my_numpy2, "bo-")
plt.subplot(1,2,2)
plt.plot(my_numpy2,my_numpy1, "g*")
```

```python
my_figure = plt.figure()
my_axes = my_figure.add_axes([.1,.1,.5,.5])
my_axes.plot(my_numpy1,my_numpy2,"r*")
my_axes.set_xlabel("X DATA")
my_axes.set_ylabel("Y DATA")
my_axes.set_title("GRAPH TITLE")
```

```python
my_figure2 = plt.figure()


my_axes3 = my_figure2.add_axes([.1,.3,.9,.9])
my_axes3.plot(my_numpy1,my_numpy2,"r--")
my_axes3.set_xlabel("X DATA LARGE")
my_axes3.set_ylabel("Y DATA LARGE")
my_axes3.set_title("LARGE-GRAPH TITLE")

my_axes2 = my_figure2.add_axes([.25,.6,.3,.3])
my_axes2.plot(my_numpy1,my_numpy2,"g*")
my_axes2.set_xlabel("X DATA SMALL")
my_axes2.set_ylabel("Y DATA SMALL")
my_axes2.set_title("SMALL-GRAPH TITLE")
```

```python
(my_fig, my_axe) = plt.subplots()
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 2
my_axe.plot(my_numpy3,my_numpy4, color="#fc0388", alpha=1)
my_axe.plot(my_numpy4,my_numpy3, color="#5a03fc", alpha=1)
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3,my_numpy3 +2 , color="#fc0388", linewidth=4, linestyle="--")
new_axe.plot(my_numpy3,my_numpy3 +3.5 , color="blue", linewidth=3, linestyle="-.")
new_axe.plot(my_numpy3,my_numpy3 +5 , color="#5a03fc", linewidth=6, linestyle=":")
new_axe.plot(my_numpy3,my_numpy3 +7 , color="#dbfc03", linewidth=2, linestyle=":", marker="o")
new_axe.plot(my_numpy3,my_numpy3 +8 , color="#03fcf0", linewidth=2, linestyle=":", marker="+", markersize=10)
```
